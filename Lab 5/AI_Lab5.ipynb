# **Case Study 1: Weather Data Analysis**

**Task 1: Data Generation Task**

```python
# Task 1: Data Generation Task
import numpy as np
import pandas as pd

# Generate random data for each column
np.random.seed(0)
dates = pd.date_range(start='2023-01-01', periods=365)
temperatures = np.random.uniform(10, 40, size=365)
humidity = np.random.uniform(30, 90, size=365)
wind_speed = np.random.uniform(0, 20, size=365)
weather_conditions = np.random.choice(['Sunny', 'Rainy', 'Cloudy'], size=365)

# Create the DataFrame
weather_data = pd.DataFrame({
    'Date': dates,
    'Temperature': temperatures,
    'Humidity': humidity,
    'Wind Speed': wind_speed,
    'Weather Condition': weather_conditions
})

# Display first few rows
weather_data.head()
```

**Task 2: NumPy Task**

```python
# Convert the Temperature column to a NumPy array
temperature_array = weather_data['Temperature'].to_numpy()

# Calculate mean, median, and standard deviation of Temperature
mean_temp = np.mean(temperature_array)
median_temp = np.median(temperature_array)
std_temp = np.std(temperature_array)

# Print the results
print(f"Mean Temperature: {mean_temp:.2f}°C")
print(f"Median Temperature: {median_temp:.2f}°C")
print(f"Standard Deviation of Temperature: {std_temp:.2f}°C")
```

**Task 3: Pandas Task (Filter for Sunny days with temperature > 30°C)**

```python
# Filter for days when Temperature > 30°C and Weather Condition is 'Sunny'
sunny_hot_days = weather_data[(weather_data['Temperature'] > 30) & (weather_data['Weather Condition'] == 'Sunny')]

# Count the number of such days
num_sunny_hot_days = sunny_hot_days.shape[0]

# Print the result
print(f"Number of Sunny Days with Temperature > 30°C: {num_sunny_hot_days}")
```

**Task 4: Pandas Task (Group by Weather Condition and calculate average Humidity)**

```python
# Group by Weather Condition and calculate average Humidity
avg_humidity_by_condition = weather_data.groupby('Weather Condition')['Humidity'].mean()

# Print the results
print(avg_humidity_by_condition)
```

**Task 5: Matplotlib Task (Plot temperature variation over the year)**

```python
import matplotlib.pyplot as plt

# Plot temperature variation over the year
plt.figure(figsize=(10, 6))
plt.plot(weather_data['Date'], weather_data['Temperature'], color='orange', label='Temperature')
plt.title('Temperature Variation Over the Year')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.legend()
plt.show()
```

**Task 6: Matplotlib Task (Bar plot of number of days for each weather condition)**

```python
# Count the number of days for each weather condition
weather_counts = weather_data['Weather Condition'].value_counts()

# Plot a bar chart
plt.figure(figsize=(6, 4))
weather_counts.plot(kind='bar', color=['yellow', 'blue', 'gray'])
plt.title('Number of Days for Each Weather Condition')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Days')
plt.show()
```

# **Case Study 2: Sales Data Analysis**

**Task 1: Data Generation Task**

```python
import numpy as np
import pandas as pd

# Set seed for reproducibility
np.random.seed(0)

# Generate dummy data
order_ids = np.arange(1, 501)
products = np.random.choice(['Product_A', 'Product_B', 'Product_C', 'Product_D', 'Product_E',
                             'Product_F', 'Product_G', 'Product_H', 'Product_I', 'Product_J'], size=500)
prices = np.random.uniform(10, 1000, size=500)
quantities = np.random.randint(1, 21, size=500)
dates_of_purchase = pd.to_datetime(np.random.choice(pd.date_range('2023-01-01', '2023-12-31'), size=500))

# Create DataFrame
order_data = pd.DataFrame({
    'Order ID': order_ids,
    'Product': products,
    'Price': prices,
    'Quantity': quantities,
    'Date of Purchase': dates_of_purchase
})

# Display the first few rows
order_data.head()
```

**Task 2: NumPy Task**

```python
# Create NumPy array containing the Price and Quantity columns
price_quantity_array = order_data[['Price', 'Quantity']].to_numpy()

# Calculate total sales for each order (Price * Quantity)
total_sales = np.prod(price_quantity_array, axis=1)

# Add total sales as a new column in the DataFrame
order_data['Total Sales'] = total_sales

# Display the first few rows with Total Sales
order_data.head()
```

**Task 3: Pandas Task (Filter orders with total sales > $100)**

```python
# Filter orders where Total Sales > $100
high_value_orders = order_data[order_data['Total Sales'] > 100]

# Display filtered data
high_value_orders.head()
```

**Task 4: Pandas Task (Group by Product and calculate total quantity sold)**

```python
# Group by Product and calculate the total quantity sold for each product
total_quantity_by_product = order_data.groupby('Product')['Quantity'].sum()

# Display the total quantity sold for each product
total_quantity_by_product
```

**Task 5: Matplotlib Task (Scatter plot of Price vs Quantity)**

```python
import matplotlib.pyplot as plt

# Create a scatter plot to visualize the relationship between Price and Quantity
plt.figure(figsize=(8, 6))
plt.scatter(order_data['Price'], order_data['Quantity'], color='blue', alpha=0.5)
plt.title('Scatter Plot of Price vs Quantity Sold')
plt.xlabel('Price ($)')
plt.ylabel('Quantity Sold')
plt.grid(True)
plt.show()
```

**Task 6: Matplotlib Task (Histogram of total sales values)**

```python
# Create a histogram to show the distribution of Total Sales values
plt.figure(figsize=(8, 6))
plt.hist(order_data['Total Sales'], bins=20, color='green', edgecolor='black', alpha=0.7)
plt.title('Histogram of Total Sales')
plt.xlabel('Total Sales ($)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
